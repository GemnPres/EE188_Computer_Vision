```python
# Write your answer in this cell. Then copy paste the code into the overleaf file corresponding to Question 2.
from math import floor

def conv2D(image: np.array, kernel: np.array = None):
    zp_rows = np.zeros((floor(kernel.shape[0]/2), image.shape[1]))
    zp_image = np.vstack((zp_rows, image, zp_rows))
    zp_cols = np.zeros((zp_image.shape[0], floor(kernel.shape[1]/2)))
    zp_image = np.hstack((zp_cols, zp_image, zp_cols))
    
    #flip h around both x and y axis
    h_flip = np.flip(kernel, 0)
    h_flip = np.flip(h_flip, 1)
    
    # N_x_n1 + N_h_n1 <= n1 <= N_x_n1 + N_h_n1 + L_x_n1 + L_h_n1 -2
    # N_x_n2 + N_h_n2 <= n2 <= N_x_n2 + N_h_n2 + L_x_n2 + L_h_n2 -2
    L_x_n1 = image.shape[1]
    L_x_n2 = image.shape[0]
    L_h_n1 = kernel.shape[1]
    L_h_n2 = kernel.shape[0]
    y = np.zeros((L_x_n2, L_x_n1))


    for n2 in range(L_x_n2):
        for n1 in range(L_x_n1):
            image_overlap = zp_image[n2:n2+L_h_n2, n1:n1+L_h_n1]
            y_n2_n1 = np.sum(np.multiply(image_overlap, h_flip)) # Convolution: elementwise multiply the two matrices and sum over all entries
            y[n2, n1] = y_n2_n1
     
    return y
```

### Gaussian Filter



The formula for a 2-D isotropic gaussian distribution with variance $\sigma^2$ and mean $= [\mu_x, \mu_y]$ is given by 

$p(x,y) = \dfrac{1}{2\pi\sigma^2}\exp{\big(-\dfrac{(x-\mu_x)^2+(y-\mu_y)^2}{2\sigma^2}\big)}$

Using the equation above, complete the function `gaussian_filter(size, sigma)` which given a filter size and standard deviation, returns a centered gaussian filter. Unlike the usual 2-D gaussian which is defined on a continuous space, for the gaussian filter you will assume that $x,y,\mu_x,\mu_y$ are discrete integers.

*Note:* Dont forget to normalize the filter to make sure that it sums to 1. Since the filter size is odd, assume the center pixel to the origin and the mean of the gaussian.

### **Answer 3.1:**
Copy paste your solution in the cell below on overleaf for Question 3.1.

```python
# Write your answer in this cell.
from math import pi, floor, exp

def gaussian_filter(size: int, sigma: float):
    mean = np.array([floor(size/2), floor(size/2)])
    p_xy_mat = np.zeros((size, size))
    
    for x in range(size):
        for y in range(size):
            p_xy = (1/(2*pi*(sigma**2)))*\
            exp(-((x-mean[1])**2 + (y-mean[0])**2)/(2*(sigma**2)))
            p_xy_mat[x, y] = p_xy
    
    p_xy_mat = (1/np.sum(p_xy_mat))*p_xy_mat
    return p_xy_mat

gauss_filt = gaussian_filter(21,5)
plt. pcolormesh(gauss_filt)
```

### Median filtering

Complete the function `median_filtering(image, kernel_size)` which takes the `image` as input along with the `kernel_size` and the returns the median filtered output which has the same size as the input image (you need to perform zero padding).

### **Answer 3.4:**
Copy paste your solution in the cell below on overleaf for Question 3.4.

```python
# Write your answer in this cell. Then copy paste the code into the overleaf file corresponding to Question 3 (d).

from math import floor

def median_filtering(image: np.array, kernel_size: int = None):
    zp_rows = np.zeros((floor(kernel_size/2), image.shape[1]))
    zp_image = np.vstack((zp_rows, image, zp_rows))
    zp_cols = np.zeros((zp_image.shape[0], floor(kernel_size/2)))
    zp_image = np.hstack((zp_cols, zp_image, zp_cols))
    
    L_x_n1 = image.shape[1]
    L_x_n2 = image.shape[0]
    y = np.zeros((L_x_n2, L_x_n1))

    for n2 in range(L_x_n2):
        for n1 in range(L_x_n1):
            image_overlap = zp_image[n2:n2+kernel_size, n1:n1+kernel_size]
            y_n2_n1 = np.median(image_overlap) # Convolution: elementwise multiply the two matrices and sum over all entries
            y[n2, n1] = y_n2_n1

    return y
```

In this question you will be completing a function `filtering_2(image, kernel, sigma_int, norm_fac)` which takes as input the `image`, a gaussian filter `kernel` which is the filter for the spatial dimension, `sigma_int` is the standard deviation of the gaussina along the intensity/pixel dimension and `norm_fac` is the normalization factor.

*Note:* For this filter you have two types of gaussian filters (one along the spatial dimension and the other along the pixel dimension). The gaussin filter along the spatial dimension can be obtained using the `gaussian_filter` function you wrote previously,however, the gaussian filter along the intensity dimension is a non-linear filter just like the median filter.

```python
# Write your code in this cell.

def filtering_2(image: np.array, kernel: np.array = None, \
                sigma_int: float = None, norm_fac: float = None):
    from math import pi, floor, exp, sqrt
    
    #kernel is symmetrical
    kernel_size = kernel.shape[0]
    kernel_center = floor(kernel_size/2)
    
    #zeropad the image
    zp_rows = np.zeros((floor(kernel.shape[0]/2), image.shape[1]))
    zp_image = np.vstack((zp_rows, image, zp_rows))
    zp_cols = np.zeros((zp_image.shape[0], floor(kernel.shape[1]/2)))
    zp_image = np.hstack((zp_cols, zp_image, zp_cols))
    
    #
    I_p_mat = np.zeros((image.shape[0], image.shape[1]))
    q_mat = np.zeros((kernel_size, kernel_size))

    for n1 in range(kernel_center, kernel_center+image.shape[1]):
        for n2 in range(kernel_center, kernel_center+image.shape[0]):
            for N1 in range(n1-kernel_center, \
                            n1-kernel_center+kernel_size):
                for N2 in range(n2-kernel_center, \
                                n2-kernel_center+kernel_size):
                    
                    I_center = zp_image[n2, n1]
                    I_q = zp_image[N2, N1]
                    
                    #intensity gaussian
                    Gauss_Intensity = \
                    (1/((sqrt(2*pi))*sigma_int))\
                    *exp(-((I_q-I_center)**2)/(2*(sigma_int**2)))
                    
                    #retrieve the spacial gaussian for neighboring positions
                    Gauss_Space = \
                    kernel[n2-N2+kernel_center, n1-N1+kernel_center]
                    
                    q_mat[n2-N2+kernel_center, n1-N1+kernel_center] = \
                    Gauss_Space*Gauss_Intensity*I_q
            
            #I_p_mat[n2-kernel_center, n1-kernel_center] = \
            #(1/norm_fac)*np.sum(q_mat)
            #norm_fac is a bit tricky here 
            #since I already normalize in the Gaussian blur. 
            #The code would print out pure white (intensity blow)
            #if divided by a small value like 0.0075
            I_p_mat[n2-kernel_center, n1-kernel_center] = np.sum(q_mat)
            q_mat = np.zeros((kernel_size, kernel_size))
                                                                  
    return I_p_mat
```

```python
gauss_filt = gaussian_filter(11,3)
gauss_filt_img_data = conv2D(img_data, gauss_filt)
bilateral_filt_img_data = filtering_2(img_data, gauss_filt, sigma_int=0.35, norm_fac=0.0075) #These are some reference values, feel free to tune the values.
```
